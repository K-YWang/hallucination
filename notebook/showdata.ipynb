# P2数据集
其中各个字段的意义为：
- Prompt：提示词
- Category：所属的类型
- Challenge：挑战类型，通常评估prompt是否容易产生幻觉
- Note：设计的目的或者特殊语义
- images：生成的图像
- model_name：使用的是什么模型
- seed：种子

```python
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset("parquet", data_files="../data/raw/p2_muse512/data/train-00000-of-00001-d56c6394e70cbcea.parquet")["train"]
print(dataset)
print(dataset[0])
```

```python
# 可视化前 N 个样本
N = 10

for i in range(N):
    sample = dataset[i]
    prompt = sample['Prompt']
    note = sample['Note']
    image = sample['images']

    plt.figure(figsize=(3, 3))
    plt.imshow(image)
    plt.axis('off')
    plt.title(f"Prompt: {prompt}\nNote: {note}", fontsize=8)
    plt.tight_layout()
    plt.show()
```
